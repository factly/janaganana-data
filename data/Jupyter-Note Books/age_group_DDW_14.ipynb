```python
file_path="/Users/mthipparthi/PycharmProjects/janaganana-data/data/c-14/DDW_14.csv"

# Reading csv file for pca data
import pandas as pd
df = pd.read_csv(file_path)
```

```python
#droping the first coumn data as it is tanle name
# df.drop("_unnamed", inplace=True, axis=1)
df.drop("C4114", inplace=True, axis=1)
df=df.dropna()
```

```python
df.columns = ['state', 'district', 'geo_level', 'age_group', 't_p', 't_m', 't_f','r_p', 'r_m', 'r_f','u_p', 'u_m', 'u_f']
df.head(5)
```

```python
df.drop(['t_m','t_f', 't_p', 'r_p', 'u_p'], inplace=True, axis=1)

df.head(5)
```

```python
# df.columns = ['geo_code', 'geo_level', 'area', 'sex', 'literacy', 'total']

def geo_level_func(ds):
    if ds.geo_level == 'India':
        ds['geo_level'] = 'country'
        ds['geo_code'] = 'IN'
    else:
#       District - South Andaman (03)
        geo_level = ds.geo_level.split('-')[0].lower()
        ds['geo_level'] = geo_level
    return ds

df=df.apply(geo_level_func, axis=1)
```

```python
df.head(5)
```

```python
def geo_code_func(ds):
    if ds.district == 0:
        ds.geo_code = ds.state
    else:
        ds.geo_code= ds.district
    return ds

df=df.apply(geo_code_func, axis=1)
```

```python
df.tail(20)
columns_you_want = ['geo_code','geo_level','age_group','r_f','r_m', 'u_f', 'u_m'] 
df=df[columns_you_want]
df.head(5)
```

```python
rural_columns_you_want = ['geo_code','geo_level','age_group','r_f','r_m'] 
df_rural=df[rural_columns_you_want]
df_rural['area']='Rural'
df_rural.rename(columns={'r_f':'F_Total'}, inplace=True)
df_rural.rename(columns={'r_m':'M_Total'}, inplace=True)
df_rural.head(5)
```

```python
urban_columns_you_want = ['geo_code','geo_level','age_group','u_f','u_m'] 
df_urban=df[urban_columns_you_want]
df_urban['area']='Urban'
df_urban.rename(columns={'u_f':'F_Total'}, inplace=True)
df_urban.rename(columns={'u_m':'M_Total'}, inplace=True)
df_urban.head(5)
```

```python
urban_female_columns_you_want = ['geo_code','geo_level','age_group','area','F_Total'] 
df_urban_f=df_urban[urban_female_columns_you_want]
df_urban_f['sex']='Female'
df_urban_f.rename(columns={'F_Total':'Total'}, inplace=True)
df_urban_f.head(5)
```

```python
urban_male_columns_you_want = ['geo_code','geo_level','age_group','area','M_Total'] 
df_urban_m=df_urban[urban_male_columns_you_want]
df_urban_m['sex']='Male'
df_urban_m.rename(columns={'M_Total':'Total'}, inplace=True)
df_urban_m.head(5)
```

```python
rural_male_columns_you_want = ['geo_code','geo_level','age_group','area','M_Total'] 
df_rural_m=df_rural[rural_male_columns_you_want]
df_rural_m['sex']='Male'
df_rural_m.rename(columns={'M_Total':'Total'}, inplace=True)
df_rural_m.head(5)
```

```python
rural_female_columns_you_want = ['geo_code','geo_level','age_group','area','F_Total'] 
df_rural_f=df_rural[rural_female_columns_you_want]
df_rural_f['sex']='Female'
df_rural_f.rename(columns={'F_Total':'Total'}, inplace=True)
df_rural_f.head(5)
```

```python
result = pd.concat([df_rural_f, df_rural_m, df_urban_f, df_urban_m])
```

```python
result.head(10)
```

```python
def geo_code_country_func(ds):
    if ds.geo_level == 'country':
        ds.geo_code = 'IN'
    return ds
result=result.apply(geo_code_country_func, axis=1)
result.head(15)
```

```python
result.tail(20)
```

```python
rs=result[result.age_group != 'All ages']
rs.tail(20)
```
